```python
import pandas as pd
from sqlalchemy import create_engine
```

### Products CSV into DataFrame

```python
csv_file = "Resource/master-item-list.csv"
product_data_df = pd.read_csv(csv_file)
product_data_df.head()
```

### New product data with select columns

```python
new_product_data_df = product_data_df[['sku', 'name', 'list_price', 'standard_dealer_price', 'brand', 'upc', 'vendor_number','product_name','product_description','product_features','primary_item_image']].copy()
new_product_data_df.head()
```

# Rename column name

```python
new_name_product_data_df = new_product_data_df.rename(columns = {'list_price': 'selling_price','standard_dealer_price':'item_cost'})
new_name_product_data_df.drop_duplicates('sku', inplace = True)
new_name_product_data_df.head()
```

### Connect to local database

```python
rds_connection_string = "postgres:Hope%401714@localhost:5432/ETL_project_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
new_name_product_data_df.to_sql(name='products', con=engine, if_exists='append', index=False)
```
